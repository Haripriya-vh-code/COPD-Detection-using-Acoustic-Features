```python
# inter SVM training

import pandas as pd
import joblib
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVC
from sklearn.metrics import classification_report, confusion_matrix

train_excel = r"D:\Projects\wetransfer_sorted-copd_2025-07-26_0604\Last_files\all_features_combined_last.xlsx"

df_train = pd.read_excel(train_excel)

df_train = df_train.loc[:, ~df_train.columns.str.contains('^Unnamed')]
df_train = df_train.dropna(subset=["label"])

X_train = df_train.drop(columns=["file_name", "label", "group"])
y_train = df_train["label"]

X_train = X_train.fillna(0)

scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)

svm_model = SVC(kernel="rbf", C=1.0, gamma="scale", probability=True, random_state=42)

svm_model.fit(X_train_scaled, y_train)

y_pred_svm = svm_model.predict(X_train_scaled)

print("SVM Classification Report:\n", classification_report(y_train, y_pred_svm))
print("SVM Confusion Matrix:\n", confusion_matrix(y_train, y_pred_svm))

train_score = svm_model.score(X_train_scaled, y_train)

print("SVM trained successfully")
print("Training Accuracy:", train_score)

print("Features used:")
print(X_train.columns.tolist())

svm_name = r"D:\Projects\wetransfer_sorted-copd_2025-07-26_0604\Last_files\inter_svm_model_copd.pkl"
joblib.dump(svm_model, svm_name)

scaler_name = r"D:\Projects\wetransfer_sorted-copd_2025-07-26_0604\Last_files\inter_svm_model_scaler_copd.pkl"
joblib.dump(scaler, scaler_name)

joblib.dump(list(X_train.columns), r"D:\Projects\inter_svm_features.joblib")
```

```python
# inter SVM testing

import pandas as pd
import joblib
from sklearn.metrics import accuracy_score, classification_report

CSV_FILE = r"D:\Projects\wetransfer_sorted-copd_2025-07-26_0604\Testing_folders\testing_feature_extracted.csv"
MODEL_PATH = r"D:\Projects\wetransfer_sorted-copd_2025-07-26_0604\Last_files\inter_svm_model_copd.pkl"
SCALER_PATH = r"D:\Projects\wetransfer_sorted-copd_2025-07-26_0604\Last_files\inter_svm_model_scaler_copd.pkl"
FEATURES_PATH = r"D:\Projects\inter_svm_features.joblib"

model = joblib.load(MODEL_PATH)
scaler = joblib.load(SCALER_PATH)
features = joblib.load(FEATURES_PATH)

print("Loaded features:", features)
df = pd.read_csv(CSV_FILE).fillna(0)
X = df[features]
X_scaled = scaler.transform(X)
df["svm_pred_label"] = model.predict(X_scaled)
df["svm_pred_prob"] = model.predict_proba(X_scaled)[:, 1]

OUTPUT_FILE = r"D:\Projects\svm_predictions.csv"
df.to_csv(OUTPUT_FILE, index=False)
print("SVM Predictions saved to:", OUTPUT_FILE)

#accuracy checking
y_true = df['true_label']
y_pred = df['svm_pred_label']

acc = accuracy_score(y_true, y_pred)

print(f"SVM Accuracy: {acc*100:.2f}%\n")

print("SVM Classification Report:\n")
print(classification_report(y_true, y_pred))
```

```python
# inter RF training

import pandas as pd
import joblib
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, confusion_matrix

train_excel = r"D:\Projects\wetransfer_sorted-copd_2025-07-26_0604\Last_files\all_features_combined_last.xlsx"
df_train = pd.read_excel(train_excel)

df_train = df_train.loc[:, ~df_train.columns.str.contains('^Unnamed')]
df_train = df_train.dropna(subset=["label"])

X_train = df_train.drop(columns=["file_name", "label", "group"])
y_train = df_train["label"]

X_train = X_train.fillna(0)
rf_model = RandomForestClassifier(n_estimators=100, random_state=42)
rf_model.fit(X_train, y_train)
y_pred = rf_model.predict(X_train)

print("Random Forest Classification Report:\n", classification_report(y_train, y_pred))
print("Random Forest Confusion Matrix:\n", confusion_matrix(y_train, y_pred))

print("Training Accuracy:", rf_model.score(X_train, y_train))
joblib.dump(rf_model, r"D:\Projects\inter_rf_model_copd.joblib")
joblib.dump(list(X_train.columns), r"D:\Projects\inter_rf_features.joblib")
```

```python
# inter RF testing

import pandas as pd
import joblib
from sklearn.metrics import accuracy_score, classification_report

CSV_FILE = r"D:\Projects\wetransfer_sorted-copd_2025-07-26_0604\Testing_folders\testing_feature_extracted.csv"
MODEL_PATH = r"D:\Projects\inter_rf_model_copd.joblib"
FEATURES_PATH = r"D:\Projects\inter_rf_features.joblib"

model = joblib.load(MODEL_PATH)
features = joblib.load(FEATURES_PATH)

print("Loaded features:", features)
df = pd.read_csv(CSV_FILE).fillna(0)
X = df[features]

# accuracy
df["rf_pred_label"] = model.predict(X)
df["rf_pred_prob"] = model.predict_proba(X)[:, 1]

OUTPUT_FILE = r"D:\Projects\rf_predictions.csv"
df.to_csv(OUTPUT_FILE, index=False)

print("Predictions saved to:", OUTPUT_FILE)
y_true = df['true_label']
y_pred = df['rf_pred_label']

acc = accuracy_score(y_true, y_pred)
print(f"RF Accuracy: {acc*100:.2f}%\n")

print("Classification Report:\n")
print(classification_report(y_true, y_pred))
```
