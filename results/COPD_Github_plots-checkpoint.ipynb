```python
# vowel quadrilateral

import matplotlib.pyplot as plt
import numpy as np

# obtained mean values from formants

data = {
    'a': {'COPD_F1': 695.19, 'Healthy_F1': 733.26, 'COPD_F2': 1374.29, 'Healthy_F2': 1353.72},
    'e': {'COPD_F1': 530.42, 'Healthy_F1': 537.41, 'COPD_F2': 1783.31, 'Healthy_F2': 1870.08},
    'i': {'COPD_F1': 490.35, 'Healthy_F1': 426.68, 'COPD_F2': 1910.53, 'Healthy_F2': 2034.27},
    'o': {'COPD_F1': 596.12, 'Healthy_F1': 549.37, 'COPD_F2': 1320.67, 'Healthy_F2': 1140.45},
    'u': {'COPD_F1': 535.48, 'Healthy_F1': 505.29, 'COPD_F2': 1345.91, 'Healthy_F2': 1120.63}
}


order = ['i', 'e', 'a', 'o', 'u']

def get_coords(group, axis):
    key = f"{group}_{axis}"
    return [data[v][key] for v in order]

h_f1 = get_coords('Healthy', 'F1')
h_f2 = get_coords('Healthy', 'F2')
c_f1 = get_coords('COPD', 'F1')
c_f2 = get_coords('COPD', 'F2')

# VSA calculation using Shoelace formula

def calculate_vsa(f2, f1):
    return 0.5 * np.abs(np.dot(f2, np.roll(f1, 1)) - np.dot(f1, np.roll(f2, 1)))

vsa_healthy = calculate_vsa(h_f2, h_f1)
vsa_copd = calculate_vsa(c_f2, c_f1)


plt.figure(figsize=(10, 8))
plt.plot(h_f2 + [h_f2[0]], h_f1 + [h_f1[0]], 'b--', alpha=0.7, label=f'Healthy (VSA: {vsa_healthy:.2f})')
plt.scatter(h_f2, h_f1, color='blue', s=100)
plt.plot(c_f2 + [c_f2[0]], c_f1 + [c_f1[0]], 'r-', alpha=0.7, label=f'COPD (VSA: {vsa_copd:.2f})')
plt.scatter(c_f2, c_f1, color='red', s=100)
for i, v in enumerate(order):
    plt.text(h_f2[i] + 10, h_f1[i], v, color='blue', fontsize=12, fontweight='bold')
    plt.text(c_f2[i] + 10, c_f1[i], v, color='red', fontsize=12, fontweight='bold')

# Invert axes to match phonetic convention
plt.gca().invert_yaxis()
plt.gca().invert_xaxis()

plt.xlabel('F2 (Hz)')
plt.ylabel('F1 (Hz)')
plt.title('Vowel Quadrilateral')
plt.legend()a
plt.grid(True, linestyle=':', alpha=0.6)
plt.savefig("vowel_quadrilateral_vsa.png", dpi=300, bbox_inches="tight")
plt.show()

print(f"Healthy VSA: {vsa_healthy:.2f}")
print(f"COPD VSA: {vsa_copd:.2f}")
```

```python
# spectral flux plot

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

file_path=r"D:\Projects\wetransfer_sorted-copd_2025-07-26_0604\Last_files\all_features_combined_last.xlsx"
df=pd.read_excel(file_path)
df.columns=df.columns.str.strip()

feature="spectral_flux"
df[feature]=pd.to_numeric(df[feature],errors='coerce')
df=df.dropna(subset=[feature,'label'])

Q1=df[feature].quantile(0.25)
Q3=df[feature].quantile(0.75)
IQR=Q3-Q1
lower_bound=Q1-1.5*IQR
upper_bound=Q3+1.5*IQR

plot_df=df[(df[feature]>=lower_bound)&(df[feature]<=upper_bound)].copy()

plt.figure(figsize=(6,5),dpi=300)
sns.set_theme(style="white")

sns.boxplot(x="label",y=feature,data=plot_df,showfliers=False,palette=["#6FA3EF","#FF6F61"],width=0.4,boxprops=dict(alpha=0.3),hue="label",legend=False)

sns.swarmplot(x="label_x",y=feature,data=plot_df,palette=["#2C5FAD","#B22222"],size=4,alpha=0.8,hue="label_x",legend=False)

plt.title("Spectral Flux Analysis",fontsize=14,fontweight='bold')
plt.ylabel("Acoustic Flux",fontsize=12)
plt.xlabel("Diagnostic Group",fontsize=12)
plt.xticks([0,1],["Healthy (0)","COPD (1)"])
sns.despine()
plt.tight_layout()

plt.savefig("Flux_Final_600DPI.png",dpi=600,bbox_inches='tight')
plt.savefig("Flux_Final_Vector.pdf",bbox_inches='tight')
print(f"Success: Removed {len(df)-len(plot_df)} outliers.")
print("Files saved: Flux_Final_600DPI.png and Flux_Final_Vector.pdf")
plt.show()
```

```python
# jitter shimmer plot

import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

file_path=r"D:\Projects\wetransfer_sorted-copd_2025-07-26_0604\Last_files\all_features_combined_last.xlsx"
df_voice=pd.read_excel(file_path)

sns.set_context("paper",font_scale=1.5)
sns.set_style("whitegrid")

features=["Jitter_local","Shimmer_local"]

df_melted=df_voice.melt(id_vars="group",value_vars=features,var_name="Feature",value_name="Value")

fig,ax=plt.subplots(figsize=(10,6))

sns.swarmplot(x="Feature",y="Value",hue="group",data=df_melted,palette=["#FF6F61","#6FA3EF"],size=5,dodge=True,hue_order=["Healthy","COPD"])

plt.title("Diagnostic Comparison of Vocal Instability",fontweight='bold',pad=20)
plt.ylabel("Value (%)",fontweight='bold')
plt.xlabel("")
plt.xticks([0,1],["Local Jitter","Local Shimmer"])

sns.despine()
plt.savefig("Voice_Analysis_HighRes.png",dpi=300,bbox_inches='tight'); plt.savefig("Voice_Analysis_Vector.pdf",bbox_inches='tight')
plt.show()
```
