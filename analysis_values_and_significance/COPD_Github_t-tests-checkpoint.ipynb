```python
import pandas as pd
from scipy.stats import ttest_ind

file_path =  r"D:\Projects\wetransfer_sorted-copd_2025-07-26_0604\Last_files\all_features_combined_last.xlsx"

df = pd.read_excel(file_path)

df['Group'] = 'Healthy'

df.iloc[:236, df.columns.get_loc('Group')] = 'COPD' # since my csv file contained 236 rows of COPD and the others are healthy
features = [

    'f1_f2_ratio',
    'f2_f3_ratio',
    'f1_f3_ratio',
    'SHR',
    'breathing_energy',
    'Jitter_local',
    'Shimmer_local',
    'spectral_flux'

]

#t-test analysis

for col in features:

    try:

        # COPD data
        copd_data = df[df['Group'] == 'COPD'][col].dropna()

        # Healthy data
        healthy_data = df[df['Group'] == 'Healthy'][col].dropna()


        copd_mean = copd_data.mean()
        copd_std = copd_data.std()
        copd_median = copd_data.median()

        healthy_mean = healthy_data.mean()
        healthy_std = healthy_data.std()
        healthy_median = healthy_data.median()


        t_stat, p_value = ttest_ind(
            copd_data,
            healthy_data,
            equal_var=False
        )



        print("\n======================================")
        print(f"Feature: {col}")
        print("======================================")

        print(f"COPD Mean     : {copd_mean:.4f}")
        print(f"Healthy Mean  : {healthy_mean:.4f}")

        print(f"T-test Value  : {t_stat:.4f}")
        print(f"P-value       : {p_value:.8f}")

    except Exception as e:

        print(f"\nError processing feature: {col}")

        print(e)
```
