```python
# jitter and shimmer extraction

import os
import numpy as np
import parselmouth

BASE_PATH = r"D:\Projects\wetransfer_sorted-copd_2025-07-26_0604"

# Store results
jitter_features = {"group": [],"label" :[], "file_name": [], "Jitter_local": [], "Shimmer_local": []}

def extract_voice_features(snd):
  
    try:
        # Skip very short audio
        if snd.get_total_duration() < 0.3:
            return np.nan, np.nan

        # Pitch
        pitch = snd.to_pitch()
        if pitch is None or np.all(np.isnan(pitch.selected_array['frequency'])):
            return np.nan, np.nan

        # PointProcess from pitch
        point_process = parselmouth.praat.call([snd, pitch], "To PointProcess (cc)")

        # Jitter (local)
        jitter = parselmouth.praat.call(point_process, "Get jitter (local)", 0, 0, 0.0001, 0.02, 1.3)

        # Shimmer (local)
        shimmer = parselmouth.praat.call([snd, point_process], "Get shimmer (local)", 0, 0, 0.0001, 0.02, 1.3, 1.6)

        # Return as percentages
        return jitter * 100, shimmer * 100
    except Exception:
        return np.nan, np.nan


for group in ["CP", "others"]:
    group_path = os.path.join(BASE_PATH, group)

    for subject in sorted(os.listdir(group_path)):
        subject_path = os.path.join(group_path, subject)
        if not os.path.isdir(subject_path):
            continue

        for file in sorted(os.listdir(subject_path)):
            if not file.endswith(".wav"):
                continue

            file_path = os.path.join(subject_path, file)
            try:
                # Load sound
                snd = parselmouth.Sound(file_path)

                # Extract jitter and shimmer
                j_val, s_val = extract_voice_features(snd)

                # Store
                jitter_features["group"].append("COPD" if group=="CP" else "Healthy")
                jitter_features["label"].append(1 if group == "CP" else 0)
                jitter_features["file_name"].append(file)
                jitter_features["Jitter_local"].append(j_val)
                jitter_features["Shimmer_local"].append(s_val)

                # Print
                print(f"File: {file} | Jitter: {j_val} % | Shimmer: {s_val} %")

            except Exception as e:
                print("Error processing:", file_path, e)
                jitter_features["group"].append("COPD" if group=="CP" else "Healthy")
                jitter_features["file_name"].append(file)
                jitter_features["Jitter_local"].append(np.nan)
                jitter_features["Shimmer_local"].append(np.nan)


# creating dataframe
df = pd.DataFrame(jitter_features)

# saving to excel
output_path = r"D:\Projects\wetransfer_sorted-copd_2025-07-26_0604\Last_files\jitter_shimmer_features.xlsx"
df.to_excel(output_path, index=False)
print("\nJitter and Shimmer extraction complete.")
print(f"Excel file saved at:\n{output_path}")
```
