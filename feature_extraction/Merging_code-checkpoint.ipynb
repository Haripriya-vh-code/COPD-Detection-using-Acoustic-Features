```python
import pandas as pd

# load feature files
df_formants = pd.read_excel( r"D:\Projects\wetransfer_sorted-copd_2025-07-26_0604\Last_files\formant_features.xlsx")
df_flux = pd.read_excel(r"D:\Projects\wetransfer_sorted-copd_2025-07-26_0604\Last_files\spectral_flux.xlsx")
df_shr = pd.read_excel(r"D:\Projects\wetransfer_sorted-copd_2025-07-26_0604\Last_files\shr_features.xlsx")
df_jitter_shimmer = pd.read_excel(r"D:\Projects\wetransfer_sorted-copd_2025-07-26_0604\Last_files\jitter_shimmer_features.xlsx")
df_breathing = pd.read_excel(r"D:\Projects\wetransfer_sorted-copd_2025-07-26_0604\Last_files\breathing_energy_features.xlsx")


# merge all features using file_name


df_final = pd.merge(df_formants, df_flux[['file_name', 'spectral_flux']], on='file_name', how='left')
df_final = pd.merge(df_final, df_shr[['file_name', 'SHR']], on='file_name', how='left')
df_final = pd.merge(df_final, df_jitter_shimmer[['file_name', 'Jitter_local', 'Shimmer_local']], on='file_name', how='left')
df_final = pd.merge(df_final, df_breathing[['file_name', 'breathing_energy']], on='file_name', how='left')

# save final dataset

output_path = r"D:\Projects\wetransfer_sorted-copd_2025-07-26_0604\Last_files\all_features_combined_last.xlsx"
df_final.to_excel(output_path,index=False)
print("Feature combination complete.")
print(f"Final file saved at:\n{output_path}")
```
