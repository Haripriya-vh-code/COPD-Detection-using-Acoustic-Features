```python
# extracting Subharmonic-to-Harmonic Ratio (SHR)

# importing packages

import os
import librosa
import numpy as np
import pandas as pd
from tqdm import tqdm

# dataset path

base_path = r"D:\Projects\wetransfer_sorted-copd_2025-07-26_0604"

# parameters

sr1 = 8000
n_fft1 = 2048
hop_length1 = 512
bandwidth1 = 20      # ±20 Hz frequency band
maximum_harmonics = 5



def band_energy(freqs, spectrum, target_freq, bandwidth=20):

    idx = np.where(
        (freqs >= target_freq - bandwidth) &
        (freqs <= target_freq + bandwidth)
    )[0]

    if len(idx) == 0:
        return 0

    return np.sum(spectrum[idx] ** 2)



def extract_shr(y, sr):

    # use first 2.5 seconds
    y = y[: int(2.5 * sr)]

    # fundamental frequency estimation using YIN
    f0 = librosa.yin(
        y,
        fmin=50,
        fmax=500,
        sr=sr
    )

    # remove unvoiced frames
    f0_voiced = f0[~np.isnan(f0)]

    if len(f0_voiced) == 0:
        return np.nan

    # median fundamental frequency
    median_f0 = np.median(f0_voiced)

    # FFT spectrum
    spectrum = np.abs(np.fft.rfft(y))
    freqs = np.fft.rfftfreq(len(y), d=1 / sr)

    harmonic_energy = 0
    subharmonic_energy = 0

    # harmonic energy

    for n in range(1, maximum_harmonics + 1):

        harmonic_freq = n * median_f0

        if harmonic_freq < sr / 2:

            harmonic_energy += band_energy(
                freqs,
                spectrum,
                harmonic_freq,
                bandwidth1
            )

    # subharmonic frequencies

    subharmonic_list = [

        median_f0 / 2,
        (3 * median_f0) / 2,
        (5 * median_f0) / 2

    ]

    # subharmonic energy

    for sub_freq in subharmonic_list:

        if sub_freq < sr / 2:

            subharmonic_energy += band_energy(
                freqs,
                spectrum,
                sub_freq,
                bandwidth1
            )

    # avoid division by zero
    if harmonic_energy == 0:
        return np.nan

    # SHR computation
    shr = subharmonic_energy / harmonic_energy

    return shr


results = []

for group in ["CP", "others"]:

    group_path = os.path.join(base_path, group)
    for subject in tqdm(
        os.listdir(group_path),
        desc=f"Processing {group}"
    ):
        subject_path = os.path.join(group_path, subject)
        if not os.path.isdir(subject_path):
            continue
        for file in os.listdir(subject_path):
            if not file.lower().endswith(".wav"):
                continue
            file_path = os.path.join(subject_path, file)
            try:

                # load audio
                y, sr = librosa.load(file_path, sr=sr1)

                # extract SHR
                shr_value = extract_shr(y, sr)

                # store results
                results.append({

                    "group": "COPD" if group == "CP" else "Healthy",
                    "label": 1 if group == "CP" else 0,
                    "file_name": f"{group}/{subject}/{file}",
                    "SHR": shr_value

                })

                print(f"{file} -> SHR: {shr_value:.4f}")

            except Exception as e:
                print(f"\nError processing: {file_path}")
                print(e)

#dataframe

df = pd.DataFrame(results)

# saving to excel

output_path = r"D:\Projects\wetransfer_sorted-copd_2025-07-26_0604\Last_files\shr_features.xlsx"

df.to_excel(
    output_path,
    index=False
)

print("\nAnalysis complete.")

print(f"Excel file saved at:\n{output_path}")
```
