```python
# extracting formants, formant ratios and RMS energy

# importing the packages

import os
import numpy as np
import pandas as pd
import librosa
import parselmouth

# parameters and dataset location

base_path =  r"D:\Projects\wetransfer_sorted-copd_2025-07-26_0604"

sr1 = 16000

frame_size_ms = 25
hop_size_ms = 10

# extracting F1, F2, F3 and ratios

def extract_formants(file_path):

    y, sr = librosa.load(file_path, sr=sr1)

    sound = parselmouth.Sound(y, sr)

    frame_size = frame_size_ms / 1000.0
    hop_size = hop_size_ms / 1000.0

    duration = sound.get_total_duration()

    formant = sound.to_formant_burg(

        time_step=hop_size,

        max_number_of_formants=5,

        maximum_formant=5500,

        window_length=frame_size,

        pre_emphasis_from=50

    )

    f1_vals = []
    f2_vals = []
    f3_vals = []

    t = 0.0

    while t < duration:

        f1 = formant.get_value_at_time(1, t)
        f2 = formant.get_value_at_time(2, t)
        f3 = formant.get_value_at_time(3, t)

        if (

            f1 is not None and
            f2 is not None and
            f3 is not None

        ):

            f1_vals.append(f1)
            f2_vals.append(f2)
            f3_vals.append(f3)

        t += hop_size

    # handling empty values

    if len(f1_vals) == 0:

        return (

            np.nan,
            np.nan,
            np.nan,
            np.nan,
            np.nan,
            np.nan

        )

    # calculating mean values

    f1_mean = np.mean(f1_vals)
    f2_mean = np.mean(f2_vals)
    f3_mean = np.mean(f3_vals)

    # calculating ratios

    f1_f2_ratio = (
        f1_mean / f2_mean
        if f2_mean != 0 else np.nan
    )

    f2_f3_ratio = (
        f2_mean / f3_mean
        if f3_mean != 0 else np.nan
    )

    f1_f3_ratio = (
        f1_mean / f3_mean
        if f3_mean != 0 else np.nan
    )

    return (

        f1_mean,
        f2_mean,
        f3_mean,

        f1_f2_ratio,
        f2_f3_ratio,
        f1_f3_ratio

    )

# identifying COPD and Healthy samples

data = []

for group in ["CP", "others"]:

    group_path = os.path.join(base_path, group)

    for subject in os.listdir(group_path):

        subject_path = os.path.join(group_path, subject)

        if os.path.isdir(subject_path):

            for file in os.listdir(subject_path):

                if file.lower().endswith(".wav"):

                    file_path = os.path.join(
                        subject_path,
                        file
                    )

                    try:

                        (
                            f1_mean,
                            f2_mean,
                            f3_mean,

                            f1_f2_ratio,
                            f2_f3_ratio,
                            f1_f3_ratio

                        ) = extract_formants(file_path)

                        # storing values

                        data.append({

                            "group": (
                                "COPD"
                                if group == "CP"
                                else "Healthy"
                            "label": 1 if group == "CP" else 0,
                            ),

                            "file_name":
                            f"{group}/{subject}/{file}",

                            "F1_mean": f1_mean,
                            "F2_mean": f2_mean,
                            "F3_mean": f3_mean,

                            "F1_F2_ratio":
                            f1_f2_ratio,

                            "F2_F3_ratio":
                            f2_f3_ratio,

                            "F1_F3_ratio":
                            f1_f3_ratio,

                        })

                        print(
                            f"Processed: {file}"
                        )

                    except Exception as e:

                        print(
                            "Error processing:",
                            file_path
                        )

                        print(e)

# creating dataframe

df = pd.DataFrame(data)

# saving to excel

output_path = r"D:\Projects\wetransfer_sorted-copd_2025-07-26_0604\Last_files\formant_features.xlsx"

df.to_excel(
    output_path,
    index=False
)

print("\nAnalysis complete.")

print(f"Excel file saved at:\n{output_path}")
```
