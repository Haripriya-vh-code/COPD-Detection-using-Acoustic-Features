```python
# extracting spectral flux

# importing packages

import os
import librosa
import numpy as np
import pandas as pd
from tqdm import tqdm

# dataset path

base_path = r"D:\Projects\wetransfer_sorted-copd_2025-07-26_0604"

# parameters

sr1 = 16000
n_fft1 = 2048
hop_length1 = 512
window_length1 = 2048

# spectral flux extraction function

def extract_spectral_flux(y):

    # STFT magnitude spectrum
    S = np.abs(
        librosa.stft(
            y,
            n_fft=n_fft1,
            hop_length=hop_length1,
            win_length=window_length1
        )
    )

    # frame-wise normalization
    S = librosa.util.normalize(S, axis=0)

    # difference between consecutive frames
    delta_S = np.diff(S, axis=1)

    # squared spectral difference
    flux = np.sum(delta_S ** 2, axis=0)

    # mean spectral flux
    mean_flux = np.mean(flux)

    return mean_flux

# storing results
results = []

# processing COPD and Healthy folders
for group in ["CP", "others"]:
    group_path = os.path.join(base_path, group)
    for subject in tqdm(
        os.listdir(group_path),
        desc=f"Processing {group}"
    ):

        subject_path = os.path.join(group_path, subject)
        if not os.path.isdir(subject_path):
            continue
        for file in os.listdir(subject_path):
            if not file.lower().endswith(".wav"):
                continue
            file_path = os.path.join(subject_path, file)
            try:

                # loading audio
                y, sr = librosa.load(file_path, sr=sr1)

                # extracting spectral flux
                mean_flux = extract_spectral_flux(y)

                # saving results
                results.append({

                    "group": "COPD" if group == "CP" else "Healthy",
                    "label": 1 if group == "CP" else 0,
                    "file_name": f"{group}/{subject}/{file}",
                    "spectral_flux": mean_flux

                })

                print(f"{file} -> Spectral Flux: {mean_flux:.4f}")

            except Exception as e:

                print(f"\nError processing: {file_path}")
                print(e)

# creating dataframe
df = pd.DataFrame(results)

# saving to excel

output_path = r"D:\Projects\wetransfer_sorted-copd_2025-07-26_0604\Last_files\spectral_flux.xlsx"

df.to_excel(
    output_path,
    index=False
)

print("\nAnalysis complete.")

print(f"Excel file saved at:\n{output_path}")
```
