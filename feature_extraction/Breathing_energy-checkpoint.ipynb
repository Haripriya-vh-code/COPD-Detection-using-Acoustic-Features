```python
# breathing energy extraction

import os, librosa, numpy as np, pandas as pd

base_path = r"D:\Projects\wetransfer_sorted-copd_2025-07-26_0604"

sr1 = 8000
frame_length1 = int(0.025 * sr1)
hop_length1 = int(0.010 * sr1)

pause_threshold = 0.02
minimum_pause_duration = 0.2

def extract_breathing_energy(y, sr):
    rms = librosa.feature.rms(y=y, frame_length=frame_length1, hop_length=hop_length1)[0]
    times = librosa.frames_to_time(np.arange(len(rms)), sr=sr1, hop_length=hop_length1)
    rms_norm = rms / (np.max(rms) + 1e-8)
    pause_mask = rms_norm < pause_threshold

    pauses = []
    start = None
    for i, is_pause in enumerate(pause_mask):
        if is_pause and start is None:
            start = i
        elif not is_pause and start is not None:
            end = i
            pause_duration = times[end - 1] - times[start]
            if pause_duration >= minimum_pause_duration:
                pauses.append((start, end))
            start = None

    if start is not None:
        end = len(pause_mask) - 1
        pause_duration = times[end] - times[start]
        if pause_duration >= minimum_pause_duration:
            pauses.append((start, end))

    breathing_energy = []

    for start_idx, end_idx in pauses:
        prev_idx = max(start_idx - 1, 0)
        next_idx = min(end_idx, len(rms) - 1)
        energy = rms[prev_idx] + rms[next_idx]
        breathing_energy.append(energy)

    mean_breathing_energy = np.mean(breathing_energy) if len(breathing_energy) > 0 else 0
    return mean_breathing_energy

breathing_features = {"group": [],"label": [], "file_name": [], "breathing_energy": []}

for group in ["CP", "others"]:
    group_path = os.path.join(base_path, group)
    for subject in sorted(os.listdir(group_path)):
        subject_path = os.path.join(group_path, subject)
        if not os.path.isdir(subject_path):
            continue
        for file in sorted(os.listdir(subject_path)):
            if not file.endswith(".wav"):
                continue
            file_path = os.path.join(subject_path, file)

            try:
                y, sr = librosa.load(file_path, sr=sr1)
                breathing_energy = extract_breathing_energy(y, sr)
                breathing_features["group"].append("COPD" if group == "CP" else "Healthy")
                breathing_features["label"].append(1 if group == "CP" else 0)
                breathing_features["file_name"].append(f"{group}/{subject}/{file}")
                breathing_features["breathing_energy"].append(breathing_energy)
                print(f"File: {file} | Breathing Energy: {breathing_energy:.6f}")

            except Exception as e:
                print("Error processing:", file_path, e)
                breathing_features["group"].append("COPD" if group == "CP" else "Healthy")
                breathing_features["file_name"].append(f"{group}/{subject}/{file}")
                breathing_features["breathing_energy"].append(np.nan)

df = pd.DataFrame(breathing_features)
output_path = r"D:\Projects\wetransfer_sorted-copd_2025-07-26_0604\Last_files\breathing_energy_features.xlsx"
df.to_excel(output_path, index=False)
print("\nBreathing energy extraction complete.")
print(f"Excel file saved at:\n{output_path}")
```
